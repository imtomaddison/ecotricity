## Calculating standing cost challenge

#### In general when pricing energy supplies the bills are made up of two elements standing charge and unit rate charge. Standing charge is the sum of all the costs that the customers have to pay for the previllage of being connected to the grid. There is also a unitrate cost which depends on the actual usage of a given site.

#### MAP prices varies from period to period and they also depend on the type of meter the site has and also it's capacity.
